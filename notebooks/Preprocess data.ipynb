```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import arff
import pandas as pd
```

```python
name = "2d-10c"

df = pd.DataFrame(arff.loadarff(f"../data/artificial/{name}.arff")[0]).sample(frac=1).reset_index(drop=True)

print(df)

df["x"] = df["a0"]
df["y"] = df["a1"]
df["target"] = df["CLASS"].map({x: i for i, x in enumerate(df["CLASS"].unique())})

plt.figure(figsize=(5, 5))
plt.scatter(df["x"], df["y"], c=df["target"], s=10, cmap="jet")
plt.savefig(f"{name}.png", dpi=100)
plt.show()

df[["x", "y", "target"]].to_csv(f"{name}.csv", index=False)
```
